```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = dfpumpkin.iloc[:, :-1].values
y = dfpumpkin.iloc[:, -1:].values
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
labelEn = LabelEncoder()
y = labelEn.fit_transform(dfpumpkin['Class'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_trainsc = scaler.fit_transform(x_train)
x_testsc = scaler.transform(x_test)
```

```python
import seaborn as sns
plt.figure(figsize = (10, 8))
correlation_matrix = dfpumpkin[['Area', 'Perimeter', 'Major_Axis_Length', 'Minor_Axis_Length', 'Convex_Area', 'Equiv_Diameter', 'Eccentricity', 'Solidity', 'Extent', 'Roundness', 'Aspect_Ration', 'Compactness']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import Perceptron
perceptron = Perceptron(random_state=0)
perceptron.fit(x_trainsc, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_prediksi = perceptron.predict(x_testsc)
akurasi_perceptron = accuracy_score(y_test, y_prediksi)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sb
```

```python
print(classification_report(y_test, y_prediksi))
sb.heatmap(confusion_matrix(y_test, y_prediksi), annot=True, fmt='d')
```

```python
feature = ['Area', 'Perimeter', 'Major_Axis_Length', 'Minor_Axis_Length', 'Convex_Area', 'Equiv_Diameter', 'Eccentricity', 'Solidity', 'Extent', 'Roundness', 'Aspect_Ration', 'Compactness']

# Pentingnya Fitur perceptron
if hasattr(perceptron, 'coef_'):
    perceptron_penting = pd.DataFrame({'Feature': feature, 'Importance': np.abs(perceptron.coef_[0])}).sort_values('Importance', ascending=False)
    print("\nPentingnya Fitur perceptron:")
    print(perceptron_penting)
```

```python
if hasattr(perceptron, 'coef_'):
    plt.figure(figsize = (6, 5))
    sns.barplot(data = perceptron_penting, x='Feature', y='Importance')
    plt.title('Pentingnya Fitur Perceptron')
    plt.xticks(rotation = 45)
    plt.tight_layout()
    plt.show()
```

```python
print("\nLaporan Klasifikasi Perceptron:")
print(classification_report(y_test, y_prediksi, target_names=['Çerçevelik', 'Ürgüp Sivrisi']))
```
